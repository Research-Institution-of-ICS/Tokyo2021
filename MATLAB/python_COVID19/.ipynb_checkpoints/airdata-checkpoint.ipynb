```python
#!/usr/bin/python


from bs4 import BeautifulSoup

import requests

import pandas as pd

def getList(url):

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/21.0.1180.71 Safari/537.1 LBBROWSER'}

    res = requests.get(url, headers=headers)

    soup = BeautifulSoup(res.content, 'lxml', from_encoding="gb18030")


    table = soup.find('table', attrs={'class': 'tablea'})
    tdbg = table.find_all('tr', attrs={'height':'25'})

    data_list = []
    for item in tdbg:
        data = item.find_all('td')
        airport_code = data[0].get_text()
        airport_code = "".join(airport_code.split())
        name = data[1].get_text()
        airport_name_E = "".join(i for i in name if ord(i) < 256) 
        airport_name_E = airport_name_E.strip()
        airport_name_C = name.replace(airport_name_E,'')
        airport_name_C = "".join(airport_name_C.split())
        country = data[2].get_text()
        country = "".join(country.split())
        index_list = []
        index_list.append(airport_code)
        index_list.append(airport_name_E)
        index_list.append(airport_name_C)
        index_list.append(country)
        data_list.append(index_list)
        print(index_list)
    return data_list


def main():
    urls = []
    for i in range(40):
        urls.append('http://www.zou114.com/sanzidaima/index.asp?zm=&page='+str(i+1))


    airport_code_list = []
    airport_name_E_list = []
    airport_name_C_list = []
    country_list = []
    for url in urls:
        lst = getList(url)
        for item in lst:
            airport_code_list.append(item[0])
            airport_name_E_list.append(item[1])
            airport_name_C_list.append(item[2])
            country_list.append(item[3])

    data_dict = {'airport_code': airport_code_list, 'airport_name_E': airport_name_E_list,
                 'airport_name_C': airport_name_C_list, 'country': country_list}
    
    data = pd.DataFrame(data_dict)
    data.to_csv(r'C:\Users\lenovo\Desktop\air_trl.csv', index=0)
    # file = pd.read_csv(r'C:\Users\tecmry\Desktop\fre_city1.csv')

if __name__ == '__main__':
    main()
```
